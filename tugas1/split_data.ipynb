```python
import pandas as pd
```

```python
data = pd.read_csv("Game_of_Thrones_Script.csv")
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
def select_col(df, col_names):
    new_df = df.loc[:, col_names]
    return new_df
```

```python
selected_col = ['Season', 'Episode Title', 'Name']
data_new1 = select_col(df, selected_col)
```

```python
selected_col = ['Release Date', 'Episode', 'Sentence']
data_new2 = select_col(df, selected_col)
```

```python
data_new1.to_csv('GOT_new1.csv',index=False)
```
